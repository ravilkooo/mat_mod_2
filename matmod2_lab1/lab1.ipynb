```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = None
with open('matmod2_lab1/out_3.txt', 'r') as f:
    arr = list(map(int,f.readline().split()))
```

```python
arr_2 = None
with open('matmod2_lab1/out_4.txt', 'r') as f:
    arr_2 = list(map(int,f.readline().split()))
```

```python
arr = np.array(arr)
arr_2 = np.array(arr_2)
min_len = min(arr.shape[0], arr_2.shape[0])
arr = arr[:min_len]
arr_2 = arr_2[:min_len]
```

```python
plt.figure(figsize=(10,5))
plt.hlines(4400,0,min_len,'k')
plt.plot(arr, 'b', label='$task_3$')
plt.plot(arr_2, 'c', label='$task_4$')
plt.plot(arr, 'b')
plt.legend()
plt.title('Размер популяции')
plt.show()
```
